```python
import pandas
import seaborn
import matplotlib.pyplot as plt

''' 
The best way to complete this activity is to copy the code in this file into a jupyter notebook file you create. This will also you to see your visualizations easier. Once you have finished creating your visuals copy your code back into this file and submit it to Replit. Don't forget to also upload your new code to Github! 
'''
# call in our titanic csv
titanic_data = pandas.read_csv("titanic.csv")
titanic_data
```

```python
# Step 1: Create a heatmap showing the correlation between the following features: survival, pclass, sex, age, and fare 

titanic_features = titanic_data[["Survived", "Pclass", "Sex", "Age", "Fare"]]

seaborn.heatmap(titanic_features.corr(), annot = True, cmap = 'rocket')
plt.show()
```

```python
# Step 2: Create a bar graph that illustrates how many passengers fell into each of these age ranges: 0 - 16, 17 - 25, 26 -40, 41 - 59, 60 or older
# age_group_chart =
# titanic_maxage = titanic_data[]

age_groups = {
    
    "groups": [
        "16 or younger",
        "17 to 25",
        "26 to 40",
        "41 to 60",
        "60+"
    ],
    
    "count": [
        len( titanic_data[ (titanic_data["Age"] >= 0) & (titanic_data["Age"] <= 16) ]),
        len( titanic_data[ (titanic_data["Age"] >= 17) & (titanic_data["Age"] <= 25) ]),
        len( titanic_data[ (titanic_data["Age"] >= 26) & (titanic_data["Age"] <= 40) ]),
        len( titanic_data[ (titanic_data["Age"] >= 41) & (titanic_data["Age"] <= 59) ]),
        len( titanic_data[ (titanic_data["Age"] >= 60 )])],
}
vis_df = pandas.DataFrame(age_groups)
seaborn.barplot(data = vis_df, x = 'groups', y = 'count')
plt.show()
```

```python
# Step 3: Create a line graph showing the average survival percentage of each of the previous age groups
# survival_age_pers = 

survival_age= {
    
    "groups": [
        "16 or younger",
        "17 to 25",
        "26 to 40",
        "41 to 60",
        "60+"
    ],
    
    "mean_survival": [

        titanic_data[ titanic_data["Age"] >= 0 & (titanic_data["Age"] <= 16) ]['Survived'].mean(),
        titanic_data[ titanic_data["Age"] >= 17 & (titanic_data["Age"] <= 25) ]['Survived'].mean(),
        titanic_data[ titanic_data["Age"] >= 26 & (titanic_data["Age"] <= 40) ]['Survived'].mean(),    
        titanic_data[ titanic_data["Age"] >= 41 & (titanic_data["Age"] <= 59) ]['Survived'].mean(),
        titanic_data[ titanic_data["Age"] >= 60 ]['Survived'].mean()]
    }

vis_df = pandas.DataFrame(survival_age)     
seaborn.lineplot(data = vis_df, x = 'groups', y = 'mean_survival')
plt.show()
```
